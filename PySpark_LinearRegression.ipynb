```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('LinReg').getOrCreate()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
alldata = spark.read.format('libsvm').load('sample.txt')
```

```python
alldata.show()
```

```python
lr = LinearRegression(featuresCol='features', labelCol = 'label', predictionCol = 'prediction')
```

```python
lrModel = lr.fit(alldata)
```

```python
print('coefficients: {}'.format(str(lrModel.coefficients)))
```

```python
print('Intercept: {}'.format(str(lrModel.intercept)))
```

```python
model_summary = lrModel.summary
```

```python
model_summary.residuals.show()
```

```python
model_summary.rootMeanSquaredError
```

```python
model_summary.r2
```

```python
train_data, test_data = alldata.randomSplit([0.8, 0.2])
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
New_Model = lr.fit(train_data)
```

```python
test_results = New_Model.evaluate(test_data)
```

```python
print('RMSE For Model is: {}'.format(test_results.rootMeanSquaredError))
```

```python
new_DATA = test_data.select('features')
```

```python
prediction = New_Model.transform(new_DATA)
```

```python
prediction.show()
```
